```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras.activations import sigmoid
```

```python
# define 2 data points
# each x_i is an array of n features
# we have only 1 feature 
x_train = np.array([[1], [2]])
y_train = np.array([[300], [500]])

fig, ax = plt.subplots()
ax.scatter(x_train, y_train)
```

At minimum, a layer can be understood as the function itself. Because we want to fit a straight line, we want a **layer** with **linear** model. In TensorFlow, we call it a **linear activation**. A layer is simply an abstraction over the model, meaning its sole responsiblity is to predict a value given a set of **weights** and **bias**. It's not concerned with learning.

```python
layer = tf.keras.layers.Dense(units=1, activation='linear')
# first we need to instantiate the weight by running the layer on a sample
a1 = layer(x_train[0].reshape(1,1))
```

```python
w, b = layer.get_weights()
print(w, b)
```

```python
# Set weights and bias for layer
# Layer has only 1 unit
layer.set_weights([
    np.array([ # weights
        [200] # unit 1
    ]),
    np.array([ # bias
        100
    ])
])

w, b=layer.get_weights()
```

Now we have defined the weights and bias for our model (activation). We can predict the value.

```python
pred = layer(x_train)
print(pred)
```

```python
print(w @ x_train[0] + b)
print(w @ x_train[1] + b)
```

We can see that the layer predicts the same value as numpy's linear model using the resuling weights and bias.

```python
fig, ax = plt.subplots()
ax.plot(x_train, pred,c='green')
ax.scatter(x_train, y_train)
```

Now let's try a sigmoid activation

```python
X_train = np.array([0., 1, 2, 3, 4, 5], dtype=np.float32).reshape(-1,1)  # 2-D Matrix
Y_train = np.array([0,  0, 0, 1, 1, 1], dtype=np.float32).reshape(-1,1)  # 2-D Matrix
```

```python
pos = Y_train == 1
neg = Y_train == 0
X_train[pos]
```

```python
fig, ax = plt.subplots()
ax.scatter(X_train[pos], Y_train[pos], marker='x', label='y=1')
ax.scatter(X_train[neg], Y_train[neg], label='y=0')

ax.legend()
```

```python
log_layer = tf.keras.layers.Dense(units=1, activation="sigmoid")
```

```python
log_model = Sequential([
    tf.keras.layers.Dense(1, input_dim = 1, activation="sigmoid", name="L1")
])
```

```python
log_model.summary()
```

```python
layer = log_model.get_layer('L1')
```

```python
w, b = layer.get_weights()
print(w, b)
```

```python
layer.set_weights([
    np.array([
        [2]
    ]),
    np.array([
        -4.5
    ])
])
```

```python
Y_pred = layer(X_train)
print(Y_pred)
```

```python
Y_pred = log_model.predict(X_train)
```

```python
Y_pred
```

```python
fig, ax = plt.subplots()
ax.scatter(X_train[pos], Y_train[pos], marker='x', label='y=1')
ax.scatter(X_train[neg], Y_train[neg], label='y=0')
ax.scatter(X_train[Y_pred >= 0.5], Y_pred[Y_pred >= 0.5])
ax.scatter(X_train[Y_pred < 0.5], Y_pred[Y_pred < 0.5])

ax.legend()
```
